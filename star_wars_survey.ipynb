```python
import pandas as pd
import numpy as np

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

star_wars.head(10)
```

```python
star_wars.columns
```

## Keep rows with valid Respondent IDs

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head()
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

## Change  "yes" and "no" values to boolean values for calculation later. It is possible to sum up boolean values

```python
yes_no = {"Yes": True, "No": False}

for c in ["Have you seen any of the 6 films in the Star Wars franchise?",
         "Do you consider yourself to be a fan of the Star Wars film franchise?",
         "Do you consider yourself to be a fan of the Star Trek franchise?"]:
    star_wars[c] = star_wars[c].map(yes_no)

star_wars.head()
```

## Clean data for 'Which of the following Star Wars films have you seen? Please select all that apply.' column to boolean and rename relevant columns in order.

```python
movies = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movies)

star_wars.head()
```

## Rename columns. seen_1 to seen_6 follow corresponds to each Star Wars episode

```python
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
})

star_wars.head()
```

## Convert ranking columns to float and rename columns

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6"
})


star_wars.head()
```

```python
movie_rankings = star_wars[star_wars.columns[9:15]]
movie_rankings.mean()

```

## Average ranking for each Star Wars movie?  

```python
import matplotlib.pyplot as plt
%matplotlib inline


plt.bar(range(1, 7), star_wars[star_wars.columns[9:15]].mean())
plt.xlabel("Star Wars Movie Episodes")
plt.ylabel("Average Ranking")
plt.title("Star Wars Ratings")
```

#### Seems like the original Star Wars movies (4, 5 and 6) are rated more highly than the new ones. Note that lower numbers for average ranking means higher preference.

## No. of people who seen each movie?

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
%matplotlib inline

plt.bar(range(1, 7), star_wars[star_wars.columns[3:9]].sum())
plt.xlabel("Star Wars Movie Episodes")
plt.ylabel("Count of viewers")
plt.title("Star Wars Viewings")
```

#### We established earlier that original movies are ranked better. As such, it makes sense that they were seen by more people as well

## How does ranking differ by Gender?

```python
star_wars['Gender'].value_counts()
```

```python
male_respondents = star_wars[star_wars['Gender'] == "Male"]
female_respondents = star_wars[star_wars['Gender'] == "Female"]

fig = plt.figure(figsize=(9,9))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.bar(range(1, 7), male_respondents[male_respondents.columns[9:15]].mean(), color= "orange")
ax2.bar(range(1, 7), female_respondents[female_respondents.columns[9:15]].mean(), color= "orange")
ax3.bar(range(1, 7), male_respondents[male_respondents.columns[3:9]].sum())
ax4.bar(range(1, 7), female_respondents[female_respondents.columns[3:9]].sum())

ax1.set_title("Male Ratings")
ax1.set_ylabel("Average Rankings")
ax2.set_title("Female Ratings")
ax3.set_title("Male Viewers")
ax4.set_title("Female Viewers")
ax3.set_ylabel("Count of Viewers")
ax3.set_xlabel("Star Wars Episode")
ax4.set_xlabel("Star Wars Episode")
```

#### From the bar charts above, it seems like overall, the original movies are still rated better regardless of gender. The difference of rankings between the old and new movies are more stark for males. For females, episode I is ranked slightly lower than episode IV, but there were more of them who watched I than IV. Also note that there were more males who watched the new Star Wars movies, even though they liked them way less than the females.

## How does ranking differ by people consider themselves as either Star Wars or Star Trek Fans?

```python
star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"].value_counts()
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts()
```

#### Some people are fans of both Star Wars and Star Trek franchises. We are going to segment this analysis into 3 groups - people who are only Star Wars fan, people who are only Star Trek fans, and both.

```python
star_w = "Do you consider yourself to be a fan of the Star Wars film franchise?"
star_t = "Do you consider yourself to be a fan of the Star Trek franchise?"

star_wars_fan = star_wars[(star_wars[star_w] == True) & (star_wars[star_t] == False)]
star_trek_fan = star_wars[(star_wars[star_w] == False) & (star_wars[star_t] == True)]
trek_wars_fan = star_wars[(star_wars[star_w] == True) & (star_wars[star_t] == True)]

fig = plt.figure(figsize=(16,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

ax1.bar(range(1, 7), star_wars_fan[star_wars_fan.columns[9:15]].mean())
ax2.bar(range(1, 7), star_trek_fan[star_trek_fan.columns[9:15]].mean())
ax3.bar(range(1, 7), trek_wars_fan[trek_wars_fan.columns[9:15]].mean())


ax1.set_title("Star War Fans")
ax1.set_ylabel("Average Rankings")
ax2.set_title("Star Trek Fans")
ax3.set_title("Both")
ax2.set_xlabel("Star Wars Episode")
```

#### People who consider themselves Star War Fans only or both prefer the original movies over the new ones. For Star Trek Fans, besides episode 5, there is no significant difference in preference between the new and original movies.
